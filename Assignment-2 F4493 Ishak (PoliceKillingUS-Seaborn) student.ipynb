## 2: Plot 12 most dangerous cities

```python
kill.city.value_counts() [0:12]
```

```python
plt.figure(figsize=(8,6))
city_dang=kill.city.value_counts() [0:12]
sns.barplot(x=city_dang[:12].index,y=city_dang[:12].values)
plt.xticks(rotation=45)
plt.title('Most dangerous cities',color = 'blue',fontsize=15)
```

```python
# with seaborn barplot

```

```python
#COUNTPLOT
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x = 'city',
              data = kill,
              order = kill['city'].value_counts() [:12].index)
for i in ax.containers:
    ax.bar_label(i, color='purple', size=11)
plt.xticks(rotation=45)
plt.title('Most dangerous cities',color = 'blue',fontsize=15)
plt.show()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
state=kill.state.value_counts() [0:20]
state
```

```python
plt.figure(figsize=(10,7))
ax=sns.barplot(x=state[:20].index,y=state[:20].values)
plt.xticks(rotation=45)
plt.title('Most dangerous states',color = 'blue',fontsize=15)
for i in ax.containers:
    ax.bar_label(i, color='green', size=10)
```

```python
# with seaborn barplot
```

```python
#COUNTPLOT
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x = 'state',
              data = kill,
              order = kill['state'].value_counts() [:20].index)
for i in ax.containers:
    ax.bar_label(i, color='green', size=10)
plt.xticks(rotation=45)
plt.title('Most dangerous cities',color = 'blue',fontsize=15)
plt.show()
```

```python
# with seaborn countplot
```
